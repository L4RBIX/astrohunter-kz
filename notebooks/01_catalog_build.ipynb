# AstroHunter KZ Phase 2: Catalog Build

This notebook builds a small real development sample of debris-disk / infrared-excess targets from public VizieR catalogs. The output is a preliminary target sample for pipeline development, not a final science catalog and not a candidate-event result.

Scientific caution: incomplete crossmatches must not be described as complete. McDonald et al. 2017 is included only as a secondary/cautionary source because evolved-star and YSO contaminants require careful filtering.

```python
from pathlib import Path
import sys

REPO_ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
SRC_DIR = REPO_ROOT / "src"
if str(SRC_DIR) not in sys.path:
    sys.path.insert(0, str(SRC_DIR))

from astrohunter.catalogs import build_target_sample, save_catalog

CATALOG_DIR = REPO_ROOT / "catalogs"
CATALOG_DIR.mkdir(parents=True, exist_ok=True)

MAX_TARGETS = 20
```

## Build Development Target Sample

The builder queries VizieR through `astroquery`, normalizes common columns, and keeps the sample small for development.

```python
targets = build_target_sample(dev=True, max_targets=MAX_TARGETS)
target_path = save_catalog(targets, CATALOG_DIR / "target_sample.csv")
target_path, targets.shape
```

## Inspect Basic Columns

```python
display_columns = [
    "target_name",
    "ra_deg",
    "dec_deg",
    "source_catalog",
    "is_secondary_source",
    "crossmatch_status",
]
targets[[column for column in display_columns if column in targets.columns]].head(20)
```

## Source Counts

```python
if "source_catalog" in targets.columns and not targets.empty:
    source_counts = targets["source_catalog"].value_counts(dropna=False)
else:
    source_counts = None
source_counts
```

## Control Sample Boundary

This notebook does not create a control sample unless a real non-disk candidate pool exists. A matched control sample must be built from real TIC/Gaia/TESS metadata with documented selection criteria; it should not be fabricated from the target catalogs.
